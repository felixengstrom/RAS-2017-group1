```python
def plotpose(pose, color='k'):
    plt.arrow(pose[0], pose[1], 0.05*math.cos(pose[2]), 0.05*math.sin(pose[2]), color=color)


def plotmap():
    for line in lab_map_plottable:
        plt.plot(line[0], line[1])     
        
def init_particles(n_particles):
    particles = numpy.zeros([n_particles, 4])
    initialdist = norm(loc=0, scale=0.1)
    initialThetadist = norm(loc=0, scale=0.1) 
    
    for i in range(n_particles):
        particles[i, :] = ([0, 0, 0, 1] 
            + numpy.array([numpy.random.rand()*2.4, numpy.random.rand()*2.4, numpy.random.rand()*2*math.pi, 0]))
    return particles

def raytrace(m, pos, angles):
    pos = numpy.array(pos)
    dists = numpy.zeros(angles.shape)
    
    for i, angle in enumerate(angles):
        dist = numpy.inf
        for line in m[:]:        
            [x1, x2] = line[0]-pos[0]
            [y1, y2] = line[1]-pos[1]
            
            ca = math.cos(pos[2]+angle)
            sa = math.sin(pos[2]+angle)

            x1new = x1*ca + y1*sa
            x2new = x2*ca + y2*sa
            y1new = y1*ca - x1*sa
            y2new = y2*ca - x2*sa
            
            x1 = x1new
            x2 = x2new
            y1 = y1new
            y2 = y2new
            
            
            if (y1*y2 < 0 
                and (((x1>0) and (x2>0))
                or (((x1>=0) or (x2>=0)) and (x1 *y2 - x2*y1)*y2 > 0))
                and min(x1, x2)<dist): 
                d = (x1 *y2 - x2*y1)/(y2-y1)
                if(d>0):
                    dist = min(dist, d)
  
        dists[i] = dist
        
    return dists

def fake_lidar(m, pos, num_sensors=16):
    angles = numpy.linspace(0, 2*numpy.pi, num=num_sensors)
    dists = raytrace(m, pos, angles + numpy.random.normal(0, 0.1, size=angles.shape))
    dists *= numpy.random.normal(1, 0.05, size=dists.shape)
    dists[dists<0] = 0.001
    return [dists, angles]
    
def particle_prob_old(m, particle, lidar_reading):
    angles = lidar_reading[1]
    distances = lidar_reading[0]
    rays = raytrace(m, particle[:3], angles)
    probs = numpy.ones(angles.shape)
    cas = numpy.cos(angles)
    sas = numpy.sin(angles)
    for i, ray in enumerate(rays):
        angle = angles[i]
        distance = distances[i]
        probs[i] *= (norm.pdf(ray, distance, 0.5*distance))
    probs.sort()
    return sum(numpy.log(probs[probs.shape[0]/5:]))

def particle_prob(m, particle, lidar_reading):
    angles = lidar_reading[1]
    distances = lidar_reading[0]
    rays = raytrace(m, particle[:3], angles)
    probs = numpy.zeros(angles.shape)
    for i, ray in enumerate(rays):
        distance = distances[i]
        var = 0.2#*distance+0.01
        probs[i] = -((ray-distance)**2)/(var*2) - numpy.log(var)
    probs.sort()
    return sum(probs )#sum((probs[probs.shape[0]/5:]))

def update_particles_weights(m, particles, lidar_reading):
    for particle in particles:
        particle[3] = particle_prob(m, particle, lidar_reading)
```

```python

def update_particles_position(particles, odom_v, odom_w):
    
    for i in range(particles.shape[0]):
        p = particles[i,:]
        rv = numpy.random.normal(loc=0, scale=0.01)
        rw = numpy.random.normal(loc=0, scale=0.1)
        sgmn = (odom_w + rw)/2 + p[2]
        particles[i,:] = p + numpy.array([(odom_v+rv)*numpy.cos(sgmn), (odom_v+rv)*numpy.sin(sgmn) , odom_w + rw , 0])

def movetrue(truepos, vt, wt):
    sigmaold = truepos[2]    
    sigmanew = wt/2 + sigmaold
    truepos_n = truepos + numpy.array([vt*numpy.cos(sigmanew), vt*numpy.sin(sigmanew) , wt])
    odom_v = vt + numpy.random.normal(loc=0, scale=0.01)
    odom_w = wt + numpy.random.normal(loc=0, scale=0.1)
    return [truepos_n, odom_v, odom_w]



particles_nn = numpy.array(particles)
particles_nn[:, 3] = 1
vt = 0.05
wt = -numpy.pi*(1/(20.*16))
truepos_n = numpy.array(truepos)
plotparticles(particles_nn)
plotpose(truepos_n, color='r')
display.clear_output()
display.display(plt.gcf())


for i in range(30):
    [truepos_n, odom_v, odom_w] = movetrue(truepos_n, vt, wt)
    update_particles_position(particles_nn, odom_v, odom_w)
    tr_n = fake_lidar(lab_map_plottable, truepos_n)
    update_particles_weights(lab_map_plottable, particles_nn, tr_n)
    plt.figure()
    particles_nn = sample_particles(particles_nn)
    particles_nn[:, 3] = 1
    plotparticles(particles_nn)
    plotpose(truepos_n, color='r')
    plotpose(numpy.mean(particles_nn[:,0:3], axis=0), color='b')
    display.clear_output()
    display.display(plt.gcf())


    
```

```python

from matplotlib.lines import Line2D

import matplotlib.animation as animation


class Scope(object):
    def __init__(self, ax, maxt=2, dt=0.02):


    def update(self, y):
        lastt = self.tdata[-1]
        if lastt > self.tdata[0] + self.maxt:  # reset the arrays
            self.tdata = [self.tdata[-1]]
            self.ydata = [self.ydata[-1]]
            self.ax.set_xlim(self.tdata[0], self.tdata[0] + self.maxt)
            self.ax.figure.canvas.draw()

        t = self.tdata[-1] + self.dt
        self.tdata.append(t)
        self.ydata.append(y)
        self.line.set_data(self.tdata, self.ydata)
        return self.line,


def emitter(p=0.03):
    'return a random value with probability p, else 0'
    while True:
        v = np.random.rand(1)
        if v > p:
            yield 0.
        else:
            yield np.random.rand(1)

# Fixing random state for reproducibility
np.random.seed(19680801)


fig, ax = plt.subplots()
scope = Scope(ax)

# pass a generator in "emitter" to produce data for the update func
ani = animation.FuncAnimation(fig, scope.update, emitter, interval=10,
                              blit=True)


plt.show()
```
